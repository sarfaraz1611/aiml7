```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
np.random.seed(0)
xx = 2 - 3 * np.random.normal(0, 1, 20)
yy = xx - 2 * (xx ** 2) + 0.5 * (xx ** 3) + np.random.normal(-3, 3, 20)
plt.scatter(xx,yy, s=10)
plt.show()
```
